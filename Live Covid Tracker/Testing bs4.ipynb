```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.mohfw.gov.in/'
response = requests.get(URL).content 
soup = BeautifulSoup(response, 'html.parser') 
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.name)
```

```python
print(soup.title.string)
```

```python
print(soup.title.parent.name)
```
